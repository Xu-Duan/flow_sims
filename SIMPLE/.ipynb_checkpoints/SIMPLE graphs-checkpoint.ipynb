```python
import csv
import numpy
from matplotlib import pyplot, cm
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
f_20 = open("SIMPLE_data_20.txt", "r")
```

```python
f_100 = open("SIMPLE_data_100.txt", "r")
```

```python
u = numpy.zeros((100, 100))
v = numpy.zeros((100, 100))
p = numpy.zeros((100, 100))

with open("SIMPLE_data_20.txt", newline = '') as data_20:
    new_value = csv.reader(data_20, delimiter='\t')
    count = 0
    dimension = 0
    for value in new_value:
        if count == 0:
            dimension = int(value[0])
        elif count < 101:
            for i, number in enumerate(value):
                if i == dimension:
                    continue
                #print(i)
                u[i,count-1] = number
        elif count < 201:
            for j, number in enumerate(value):
                if j == dimension:
                    continue
                #print(j)
                v[j,count-101] = number
        else:
            for k, number in enumerate(value):
                if k == dimension:
                    continue
                p[k,count-201] = number
        count = count + 1
```

```python
nx = 100
ny = 100
nt = 500
nit = 50
c = 1
dx = 2 / (nx - 1)
dy = 2 / (ny - 1)
x = numpy.linspace(0, 2, nx)
y = numpy.linspace(0, 2, ny)
X, Y = numpy.meshgrid(x, y)

rho = 1
nu = .1
dt = .001
```

```python
fig = pyplot.figure(figsize=(11, 7), dpi=100)
pyplot.contourf(X, Y, p, alpha=0.5, cmap=cm.viridis)
pyplot.colorbar()
pyplot.contour(X, Y, p, cmap=cm.viridis)
pyplot.streamplot(X, Y, u, v)
pyplot.xlabel('X')
pyplot.ylabel('Y');
```
